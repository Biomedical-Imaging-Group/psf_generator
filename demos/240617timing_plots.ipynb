```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cartesian_cpu = np.load('../data/cpu_time_cartesian.npy')
polar_cpu = np.load('../data/cpu_time_polar.npy')
cartesian_gpu = np.load('../data/gpu_time_cartesian.npy')
polar_gpu = np.load('../data/gpu_time_polar.npy')
print(cartesian_cpu.shape)
```

```python
n_points = 9
n_pix_pupils = np.logspace(5, 13, n_points, base=2)
```

```python
plt.loglog(n_pix_pupils, cartesian_cpu, label='Cartesian CPU', marker='o')
plt.loglog(n_pix_pupils, polar_cpu, label='Polar CPU', marker='o')
plt.loglog(n_pix_pupils, cartesian_gpu, label='Cartesian GPU', marker='o')
plt.loglog(n_pix_pupils, polar_gpu, label='Polar GPU', marker='o')
plt.xlabel('Size of pupil')
plt.ylabel('Time (s)')
plt.legend()
plt.show()
```
