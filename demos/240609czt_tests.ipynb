# FFT2 with a disk pupil

```python
# create a mask with the form of a disk
size = 100
radius = 0.4
k = torch.linspace(-1, 1, size)
kx, ky = torch.meshgrid(k, k, indexing='ij')
mask = (kx**2 + ky**2 <= radius**2).float()
plt.imshow(mask)
plt.show()
```

```python
y = fft2(torch.fft.fftshift(mask))
y = torch.fft.fftshift(y)
plt.figure()
plt.subplot(121)
plt.imshow(torch.abs(y)**2)
plt.subplot(122)
plt.imshow(torch.angle(y))
plt.show()
```

```python
y = czt2d(torch.fft.fftshift(mask))
y = torch.fft.fftshift(y)
plt.figure()
plt.subplot(121)
plt.imshow(torch.abs(y)**2)
plt.subplot(122)
plt.imshow(torch.angle(y))
plt.show()
```

```python
y = custom_fft2(mask, k_start=-np.pi/16, k_end=np.pi/16)

plt.figure()
plt.subplot(121)
plt.imshow(torch.abs(y)**2)
plt.subplot(122)
plt.imshow(torch.angle(y))
plt.colorbar()
plt.show()
```

```python
y = custom_ifft2(mask, k_start=-np.pi/4, k_end=np.pi/4)

plt.figure()
plt.subplot(121)
plt.imshow(torch.abs(y)**2)
plt.subplot(122)
plt.imshow(torch.angle(y))
plt.colorbar()
plt.show()
```
